##Analiza danych

```python
# import bibliotek
import pandas as pd
import numpy as np
```

```python
# wczytywanie danych
gratka_df = pd.read_csv("gratka_dane_new2022.csv")
```

```python
pd.set_option('max_colwidth', 400)
gratka_df
```

```python
gratka_df.info()
```

```python
gratka_df.describe()
```

```python
#sprawdzenie duplikatów
gratka_df[gratka_df.duplicated()]
```

```python
gratka_df["area_m2"].value_counts()
```

```python
gratka_df["n_rooms"].value_counts()
```

```python
gratka_df["n_floor"].value_counts()
```

```python
gratka_df["price_per_m2"].value_counts()
```

```python
# średnia cena m2 mieszkania wg. metrażu
gratka_df.groupby("area_m2").agg(dict(price_per_m2 = "mean")).sort_values(by = "price_per_m2", ascending = False).round(decimals=2)
```

```python
# średnia cena m2 mieszkania wg. liczby pokoi
gratka_df.groupby("n_rooms").agg(dict(price_per_m2 = "mean")).sort_values(by = "price_per_m2", ascending = False).round(decimals=2)
```

```python
# średnia cena m2 mieszkania wg. nr piętra, na którym mieszkanie się znajduje
gratka_df.groupby("n_floor").agg(dict(price_per_m2 = "mean")).sort_values(by = "price_per_m2", ascending = False).round(decimals=2)
```

```python
# analiza ceny wg. metrażu
import plotly.express as px

px.box(gratka_df, 
       x = "area_m2", 
       y = "price_per_m2",
       title = "Cena 1m2 mieszkania w zależności od metrażu calego mieszkania")
```

```python
#oferty z powierzchnią większą niż 150 m2
gratka_df.query("area_m2 >= 150 ")
```

```python
#liczba ofert o powierzchni większej niż 150m2
gratka_df.query("area_m2 >= 150 ").count()
```

```python
# analiza ceny wg. ilości pokojów
import plotly.express as px

px.box(gratka_df, 
       x = "n_rooms", 
       y = "price_per_m2",
       title = "Cena 1m2 mieszkania w zależności od ilości pokoi")
```

```python
# analiza ceny wg. nr piętra, na której znajduje się nieruchomość
import plotly.express as px

px.box(gratka_df, 
       x = "n_floor", 
       y = "price_per_m2",
       title = "Cena 1m2 mieszkania w zależności od nr piętra, na której znajduje się nieruchomość")
```

```python
# analiza ilości ofert w stosunku do liczby pokoi
chart_pie1 = px.pie(gratka_df, 
                    names = "n_rooms", 
                    title = "Ilość ofert w stosunku do liczby pokoi")
chart_pie1.show()
```

```python
# analiza ilości ofert w stosunku do nr piętra
chart_pie2 = px.pie(gratka_df,  
                    names = "n_floor", 
                    title = "Ilość ofert w stosunku do nr piętra")
chart_pie2.show()
```

```python
#oferta o maksymalnej cenie
gratka_df.query("price_per_m2 == 37702 ")
```

```python
#oferta o maksymalnej powierzchni
gratka_df.query("area_m2 == 334.2 ")
```

```python
#oznaczenie dzielnicy z kolumny "location"
gratka_df["district"] = gratka_df["location"].str.replace(",","")
gratka_df["district"] = gratka_df["district"].str.replace("Kraków","")
gratka_df["district"] = gratka_df["district"].str.replace("małopolskie","")
gratka_df["district"] = gratka_df["district"].str.strip()
```

```python
gratka_df
```

```python
# analiza ceny wg. dzielnicy
import plotly.express as px

px.box(gratka_df, 
       x = "district", 
       y = "price_per_m2",
       title = "Cena 1m2 mieszkania w zależności od dzielnicy miasta Kraków")
```
